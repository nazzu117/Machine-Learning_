```python
# Now, let's show the correlation between the dataframe's numeric data.

dfcorr = df.drop('id',axis = 1).corr()
dfcorr
```

The code calculates and displays the correlation matrix, which helps to understand the linear relationships between the numeric features in the movie data.

--> Each row and column represent a numeric feature from the DataFrame (excluding "id").

-->The value at the intersection of a row (feature) and column (feature) represents the correlation coefficient between those two features.

-->For example, the value 0.0544270 at the intersection of the "popularity" row and "vote_average" column indicates a weak positive correlation between popularity and vote average.

-->A value close to 1 indicates a strong positive correlation (as one value increases, the other tends to increase as well).
-->A value close to -1 indicates a strong negative correlation (as one value increases, the other tends to decrease).
-->A value close to 0 indicates a weak or no correlation.

In summary, the code calculates and displays the correlation matrix, which helps to understand the linear relationships between the numeric features in the movie data.

### Popularity and languages¶


Let's plot the overall popularity of the films' languages from TMDB.



```python
# Group by language and sum popularity
landf = df.groupby(by='original_language')['popularity'].sum()

# Create a bar chart
sn.barplot(x=landf.index, y=landf.values)
plt.xlabel('Original Language')
plt.ylabel('Total Popularity')
plt.title('Total Popularity by Original Language')
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns


# Group by language
language_groups = df.groupby(by='original_language')

# Create a separate line plot for each language group (optional)
for language, group_df in language_groups:
  sns.kdeplot(group_df['popularity'])  # Kernel density estimation plot

# Adjust plot
plt.xlabel('Popularity')
plt.ylabel('Density')
plt.title('Distribution of Popularity by Original Language')
plt.show()
```

```python
# Group by language and sum popularity
landf = df.groupby(by='original_language')['popularity'].sum()

# Sort by popularity
landf_sorted = landf.sort_values(ascending=False)  # Sort by popularity (high to low)

# Investigate high/low popularity languages (modify thresholds as needed)
high_popularity_threshold = 10000  
low_popularity_threshold = 500  

high_popularity_languages = landf_sorted[landf_sorted >= high_popularity_threshold].index.tolist()
low_popularity_languages = landf_sorted[landf_sorted <= low_popularity_threshold].index.tolist()

print("High Popularity Languages:", high_popularity_languages)
print("Low Popularity Languages:", low_popularity_languages)

```

```python
# Further investigation (optional)
for language in high_popularity_languages:
  # Filter movies for this language
  high_popularity_movies = df[df['original_language'] == language]
  # Explore genres for these movies (consider using groupby or value_counts)
  high_popularity_genres = high_popularity_movies['genre'].value_counts()
  print("Genres associated with high popularity language", language, ":", high_popularity_genres.head())  # Show top genres

# Repeat similar exploration for low popularity languages (optional)
```

```python
# Calculate number of movies per language
language_movie_counts = df.groupby('original_language')['original_language'].count()

# Identify languages with very few movies (modify threshold as needed)
few_movies_threshold = 20  # Adjust this threshold based on your data

few_movies_languages = language_movie_counts[language_movie_counts <= few_movies_threshold].index.tolist()

print("Languages with very few movies:", few_movies_languages)
```

### Vote count and languages¶


```python
# Group by language and calculate average/median vote count
average_vote_count_by_language = df.groupby('original_language')['vote_count'].mean()
# Alternatively, use median: median_vote_count_by_language = df.groupby('original_language')['vote_count'].median()

# Print results (example using average)
print("Average Vote Count by Language:")
print(average_vote_count_by_language)
```

```python
# Create bar chart (using average vote count)
sns.barplot(x=average_vote_count_by_language.index, y=average_vote_count_by_language.values)
plt.xlabel('Original Language')
plt.ylabel('Average Vote Count')
plt.title('Average Vote Count by Original Language')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.show()
```

```python
# Example: Investigate high vote count languages
high_vote_languages = average_vote_count_by_language.nlargest(5).index.tolist()  # Top 5 languages

for language in high_vote_languages:
  # Filter movies for this language
  high_vote_movies = df[df['original_language'] == language]
  # Explore genres or other features for these movies (e.g., value_counts)
  high_vote_genres = high_vote_movies['genre'].value_counts()
  print("Genres associated with high vote language", language, ":", high_vote_genres.head())  # Show top genres
```

Outliers are data points that fall significantly outside the typical range of your data.
When outliers exist, a simple correlation coefficient (like comparing vote count directly to popularity) might not accurately reflect the overall relationship between the two variables

Now, let’s summarize the three boxplots together:

Boxplot 1: Shows data concentrated around 0 with outliers up to 30,000.

Boxplot 2: Displays the ‘popularity’ attribute with outliers above the main plot area, y-axis ranging from 0 to 5000.

Boxplot 3: Visualizes data with a median near 4 and outliers both below and above the main box, y-axis ranging from 0 to 8.
    
These boxplots are useful for comparing distributions, central tendencies, and identifying outliers in different datasets. Each boxplot provides a visual summary of the data’s spread and central value, making it easier to spot any unusual patterns or extreme values.
